```python
import pandas as pd
vcf = pd.read_pickle("vcf_select.pkl")
c = pd.read_csv("clinical.csv").drop("Unnamed: 0", axis=1).rename(columns={"PTID":"subject"})
img = pd.read_pickle('meta_allcsv.pkl')[["img_array", "subject", "label"]]
```

```python
vcf.head()
```

```python
c.head()
```

```python
c = c.rename(columns = {"Group":"GROUP"})
```

```python
a = vcf.merge(c, on = ["subject", "GROUP"]).merge(img, on = "subject")
```

```python
a["subject"].value_counts()
```

```python
a["GROUP"].value_counts()
```

```python
a["GROUP"].value_counts()
```

```python
cols = list(set(a.columns) - set(["PTID","subject", "label", "GROUP",
                                  "RID", "ID", "Group", "Phase", "SITEID", "VISCODE", "VISCODE2", "USERDATE", "USERDATE2", "update_stamp", "DX", "Unnamed: 0"]))
```

```python
X= a[cols]
y = a["GROUP"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y)
```

```python
snp_cols = set(X_train.columns).intersection(set(vcf.columns))
```

```python
X_train_snp = X_train[snp_cols]
X_test_snp = X_test[snp_cols]
```

```python
img_cols = set(X_train.columns).intersection(set(img.columns))
print(len(img.columns))
print(len(img_cols))
X_train_img = X_train[img_cols]
X_test_img = X_test[img_cols]
```

```python
clin_cols = set(X_train.columns).intersection(set(c.columns))
print(len(c.columns))
print(len(clin_cols))
X_train_clin = X_train[clin_cols]
X_test_clin = X_test[clin_cols]
```

```python
pd.DataFrame(X_train_snp).to_pickle("X_train_snp1.pkl")
pd.DataFrame(X_test_snp).to_pickle("X_test_snp1.pkl")
pd.DataFrame(y_train).to_pickle("y_train1.pkl")
pd.DataFrame(y_test).to_pickle("y_test1.pkl")

pd.DataFrame(X_train_clin).to_pickle("X_train_clinical1.pkl")
pd.DataFrame(X_test_clin).to_pickle("X_test_clinical1.pkl")

pd.DataFrame(X_train_img).to_pickle("X_train_img1.pkl")
pd.DataFrame(X_test_img).to_pickle("X_test_img1.pkl")
```
